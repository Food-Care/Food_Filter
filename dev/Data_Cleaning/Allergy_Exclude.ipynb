## 1. 우유

#### 회피 원재료
- '우유' 글자가 들어가 있지만 알러지를 안 일으키는 원재료
  - 우유향 (확인 필요)
- '우유' 글자가 안 들어가 있어도 알러지를 일으키는 원재료
  - 탈지분유, 전지분유, 연유
  - 유청, 유청단백, 유청분말
  - 카제인, 카제인나트륨
  - 치즈, 버터, 크림, 연유, 요구르트

```python
milk_all_results = []

exclude = [
    # '우유향', '요구르트향', '크림향', '버터향', '밀크향', '버터스카피향', '치즈향', '소다향'
    '향', 
    '콩우유', '초코크림'
    ]

include_keywords = [
    '탈지분유', '전지분유', '유청단백', '유청분말', '카제인',
    '치즈', '버터', '크림', '연유', '요구르트', '밀크', '유청', '유과'
    ]

for cell in df_food['RAWMTRL_NM'].dropna():
    items = [x.strip() for x in str(cell).split(',')]
    
    for item in items:
        if any(keyword in item for keyword in include_keywords) and not any(word in item for word in exclude):
            milk_all_results.append(item)
        elif '우유' in item and not any(word in item for word in exclude):
            milk_all_results.append(item)


milk_all_results = sorted(set(milk_all_results))

print("회피 원재료\n")

for i, v in enumerate(milk_all_results, start=1):
    print(f"{i}. {v}")
```

#### 대체 가능 원료 (알레르기 안전)
- 두유
- 아몬드유
- 귀리유
- 코코넛밀크
- 캐슈넛유
- 쌀유
- 향: '우유향', '요구르트향', '크림향', '버터향', '밀크향', '버터스카피향', '치즈향', '소다향'

```python
milk_rp_results = []

alt_allow_keywords = ['두유', '아몬드유', '귀리유', '코코넛밀크', '캐슈넛유', '쌀유', '콩우유']


alt_exclude = ['호두', '향', '기름', '대두유', '노두유']

for cell in df_food['RAWMTRL_NM'].dropna():
    items = [x.strip() for x in str(cell).split(',')]
    for item in items:
        if any(k in item for k in alt_allow_keywords) and not any(ex in item for ex in alt_exclude):
            milk_rp_results.append(item)
            continue

milk_rp_results = sorted(set(milk_rp_results) - set(milk_all_results))

print("\n대체 가능 원료 (알레르기 안전)\n")
for i, v in enumerate(milk_rp_results, start=1):
    print(f"{i}. {v}")
```
